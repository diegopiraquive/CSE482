#### Column Descriptions

- **Player Name**: The name of the basketball player.
- **Salary**: The player's salary.
- **Position**: The player's position on the team.
- **Age**: The player's age.
- **Team**: The NBA team the player belongs to.
- **GP** (Games Played): The number of games the player has participated in.
- **GS** (Games Started): The number of games the player started.
- **MP** (Minutes Played): Total minutes the player has played.
- **FG** (Field Goals): Total successful field goals made.
- **FGA** (Field Goal Attempts): Total field goals attempted.
- **FG%** (Field Goal Percentage): Percentage of field goals made.
- **3P** (Three-Point Field Goals): Total successful three-point field goals made.
- **3PA** (Three-Point Field Goal Attempts): Total three-point field goals attempted.
- **3P%** (Three-Point Field Goal Percentage): Percentage of three-point field goals made.
- **2P** (Two-Point Field Goals): Total successful two-point field goals made.
- **2PA** (Two-Point Field Goal Attempts): Total two-point field goals attempted.
- **2P%** (Two-Point Field Goal Percentage): Percentage of two-point field goals made.
- **eFG%** (Effective Field Goal Percentage): Adjusted field goal percentage considering three-point field goals.
- **FT** (Free Throws): Total successful free throws made.
- **FTA** (Free Throw Attempts): Total free throws attempted.
- **FT%** (Free Throw Percentage): Percentage of free throws made.
- **ORB** (Offensive Rebounds): Total offensive rebounds.
- **DRB** (Defensive Rebounds): Total defensive rebounds.
- **TRB** (Total Rebounds): Total rebounds (offensive + defensive).
- **AST** (Assists): Total assists.
- **STL** (Steals): Total steals.
- **BLK** (Blocks): Total blocks.
- **TOV** (Turnovers): Total turnovers.
- **PF** (Personal Fouls): Total personal fouls.
- **PTS** (Points): Total points scored.
- **Total Minutes**: The cumulative minutes played by the player.
- **PER** (Player Efficiency Rating): A measure of a player's per-minute production.
- **TS%** (True Shooting Percentage): A measure of shooting efficiency that considers field goals, 3-point field goals, and free throws.
- **3PAr** (Three-Point Attempt Rate): The proportion of a player's field goal attempts that are three-pointers.
- **FTr** (Free Throw Rate): The proportion of free throw attempts per field goal attempt.
- **ORB%** (Offensive Rebound Percentage): The percentage of available offensive rebounds a player grabbed.
- **DRB%** (Defensive Rebound Percentage): The percentage of available defensive rebounds a player grabbed.
- **TRB%** (Total Rebound Percentage): The percentage of total available rebounds a player grabbed.
- **AST%** (Assist Percentage): The percentage of team field goals a player assisted while he was on the floor.
- **STL%** (Steal Percentage): The percentage of opponent possessions that end with a steal by the player.
- **BLK%** (Block Percentage): The percentage of opponent two-point field goal attempts blocked by the player.
- **TOV%** (Turnover Percentage): The percentage of turnovers per 100 plays.
- **USG%** (Usage Percentage): The percentage of team plays used by a player while he was on the floor.
- **OWS** (Offensive Win Shares): An estimate of the number of wins contributed by a player due to his offense.
- **DWS** (Defensive Win Shares): An estimate of the number of wins contributed by a player due to his defense.
- **WS** (Win Shares): An estimate of the number of wins contributed by a player.
- **WS/48** (Win Shares per 48 Minutes): An estimate of the number of wins contributed by a player per 48 minutes.
- **OBPM** (Offensive Box Plus/Minus): A box score estimate of the offensive points per 100 possessions a player contributed above a league-average player.
- **DBPM** (Defensive Box Plus/Minus): A box score estimate of the defensive points per 100 possessions a player contributed above a league-average player.
- **BPM** (Box Plus/Minus): A box score estimate of the points per 100 possessions a player contributed above a league-average player.
- **VORP** (Value Over Replacement Player): A measure of a player's contribution, calculated as the difference between a player's performance and that of a hypothetical replacement player.
